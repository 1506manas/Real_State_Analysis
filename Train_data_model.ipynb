```python
df["Property"] = df["Property"].replace(['Condo', 'Single Family', 'Three Family', 'Two Family',
       'Four Family', 'Residential'], (0,1,2,3,4,5))
```

```python
df["Residential"].replace(['Condominium', 'Detached House', 'Triplex', 'Duplex', 'Fourplex'], (0,1,2,3,4), inplace = True)
```

```python
df.sort_values(by = "Locality", axis = 0, ascending = True, inplace = True, ignore_index = True)
df.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df["Locality"] = le.fit_transform(df["Locality"])
```

```python
df.head()
```

```python
df.info()
```

```python
x = df[["Year","Locality", "Estimated Value", "Property", "Residential", "num_rooms", "carpet_area"]]
y = df[["Sale Price"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1000)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
ne = KNeighborsRegressor(n_neighbors=20)
ne.fit(x_train, y_train)
```

```python
#ne.score(x_test, y_test)*100
```

```python
from sklearn.tree import DecisionTreeRegressor
dr = DecisionTreeRegressor(criterion = 'squared_error')
dr.fit(x_train, y_train)
#dr.score(x_test, y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
#lr.score(x_test, y_test)
```
